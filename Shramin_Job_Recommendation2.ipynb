# corpus

```python
final_jobs["trade_education_city_type_of_work_experience"] = final_jobs["trade"].map(str) + " " + final_jobs["education"] +" "+ final_jobs["city"]+ " "+final_jobs['type_of_work']+" "+final_jobs['experience']
final_jobs.trade_education_city_type_of_work_experience.head()
```

```python
import re
def cleanJob(jobText):
    jobText = re.sub('http\S+\s*', ' ', jobText)  # remove URl
    jobText = re.sub('RT|cc', ' ', jobText)  # remove RT and cc
    jobText = re.sub('#\S+', '', jobText)  # remove hashtags
    jobText = re.sub('@\S+', '  ', jobText)  # remove mentions
    jobText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', jobText)  # remove punctuations
    jobText = re.sub(r'[^\x00-\x7f]',r' ', jobText) 
    jobText = re.sub('\s+', ' ', jobText)  # remove extra whitespace
    return jobText
```

```python
final_jobs["cleaned_job"] = final_jobs.trade_education_city_type_of_work_experience.apply(lambda x: cleanJob(x))
print (final_jobs["cleaned_job"])
```

```python


#converting all the characeters to lower case
final_jobs['cleaned_job'] = final_jobs['cleaned_job'].str.lower() 
final_jobs.cleaned_job.head()

```

```python
final_all = final_jobs[['job_id', 'cleaned_job']]
# renaming the column name as it seemed a bit complicated
final_all = final_jobs[['job_id', 'cleaned_job']]
final_all = final_all.fillna(" ")

final_all.head()
```

# NLTK stands for Natural Languague Toolkit¶

```python
cleaned_job = final_all['cleaned_job']
#removing stopwords and applying potter stemming
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stemmer =  PorterStemmer()
stop = stopwords.words('english')
only_text = cleaned_job.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
only_text.head()
```

```python


only_text = only_text.apply(lambda x : filter(None,x.split(" ")))
print(only_text.head())

```

```python
only_text = only_text.apply(lambda x : [stemmer.stem(y) for y in x])
print(only_text.head())
```

```python


only_text = only_text.apply(lambda x : " ".join(x))
print(only_text.head())

```

```python


#adding the featured column back to pandas
final_all['text']= only_text
# As we have added a new column by performing all the operations using lambda function, we are removing the unnecessary column
#final_all = final_all.drop("pos_com_city_empType_jobDesc", 1)

list(final_all)
final_all.head()

```

# Using vector space model ( Cosine similarity )¶

```python
#creating tf-idf of user query and computing cosine similarity of user with job corpus
from sklearn.metrics.pairwise import cosine_similarity
candidate_tfidf = tfidf_vectorizer.transform(user_q['cleaned_candidate'])

#job_tfidf from above
tfidf_jobid

output = map(lambda x: cosine_similarity(candidate_tfidf, x),tfidf_jobid)
output
```

```python
output2 = list(output)
output2
```

```python
#getting the job id's of the recommendations
top = sorted(range(len(output2)), key=lambda i: output2[i], reverse=True)[:3]
top
recommendation = pd.DataFrame(columns = ['CandidateID', 'JobID'])
count = 0
for i in top:
    recommendation.loc[count, 'CandidateID']=u
    recommendation.loc[count,'JobID']=final_all['job_id'][i]
    count += 1
```

```python
#getting the job ids and their data
nearestjobs = recommendation['JobID'][:10]
job_description = pd.DataFrame(columns = ['JobID','text'])
for i in nearestjobs:
    index = np.where(final_all['job_id'] == i)[0][0]    
    job_description.loc[count, 'CandidateID']=u
    job_description.loc[count, 'JobID']=i
    job_description.loc[count, 'text']= final_all['text'][index]
    count += 1
    
```
